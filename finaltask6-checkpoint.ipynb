```python
dateparse = lambda dates: datetime.strptime(dates, '%m/%d/%Y')
data = pd.read_csv('retail sale.csv', parse_dates=['DATE'], index_col='DATE',date_parser=dateparse)
```

```python
data
```

```python
data.index
```

```python
x=data['RSAFSNA']
```

```python
x['2017':]
```

```python
plt.plot(x)
```

```python
#there is overall incresing trend so lets using decompositoin 
from pylab import rcParams
import statsmodels.api as sm
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(x, model='additive')
fig = decomposition.plot()
plt.show()
```

```python
#The plot above clearly shows that the sales trend is unstationary as it decreases, along with its obvious seasonality.
```

```python
#lets fit arima model
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(x, order=(1, 1, 1))  
results_ARIMA = model.fit()  
print(results_ARIMA.aic)
```

```python
print(results_ARIMA.summary().tables[1])
```

```python
pred = results_ARIMA.predict(start=pd.to_datetime('2017-01-01'), dynamic=False)
pred_ci = pred.conf_int()
ax = x['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
plt.show()
```

```python
data['forecast'] = results_ARIMA.predict(start = 250, end= 324, dynamic= True)  
data[['RSAFSNA', 'forecast']].plot(figsize=(20, 8))
```
